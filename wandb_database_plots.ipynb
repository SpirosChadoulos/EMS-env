# Creating Plots based on the collected experiment data in our WandB database

## Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Reward mean figure - Resistant - PPO vs A2C - *100

```python
data = pd.read_csv('data_csv/a2c_vs_ppo_resistant.csv')
# Remove unnecessary columns
data = data.iloc[:,[1,4]]
# Rename data columns
data.columns = ['PPO','A2C']
data
```

```python
marker_every = 10
a2c_color = 'green'
ppo_color = 'blue'

# Create a plot and specify x and y
columns_to_plot = {
    'PPO': {'style': '-', 'marker': ',', 'markevery': marker_every, 'color': ppo_color, 'linewidth': 4},
    'A2C': {'style': '--', 'marker': ',', 'markevery': marker_every, 'color': a2c_color, 'linewidth': 4},
}

ax = data.plot(y=columns_to_plot.keys(), figsize=(20, 20))

# Increase font size for axis labels
ax.set_xlabel('Training Episode', fontsize=35, labelpad=25)
ax.set_ylabel('Episode Reward Mean', fontsize=35, labelpad=25)

# Increase font size for tick labels
ax.tick_params(axis='x', which='major', labelsize=32)
ax.tick_params(axis='y', which='major', labelsize=32)
ax.yaxis.offsetText.set_fontsize(24)

# Remove scientific notation
# ax.ticklabel_format(style='plain')


# Apply custom styles to each line in the plot
lines = ax.get_lines()
for col, style in columns_to_plot.items():
    line = lines[list(columns_to_plot.keys()).index(col)]
    line.set_linestyle(style['style'])
    line.set_marker(style['marker'])
    line.set_markevery(style['markevery'])
    line.set_color(style['color'])
    line.set_linewidth(style['linewidth'])
    #line.set_markersize(50)

# Customize the legend
legend = ax.legend(loc=2, prop={'size': 25}) 

plt.savefig('paper_plots/resistant_a2c_vs_ppo.png')
plt.show()
```
